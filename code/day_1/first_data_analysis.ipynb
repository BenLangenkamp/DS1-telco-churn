Skewness measures the asymmetry of a distribution's probability distribution:

Positive skew: When the right tail is longer (data is concentrated on the left)

Zero skew: Symmetric distribution

Negative skew: When the left tail is longer (data is concentrated on the right)

```python
print("Mathenote skewness: ", df_clean.mathe_note.skew())
```

Kurtosis measures the "tailedness" of a distribution:

Higher kurtosis: Heavy tails, more outliers (leptokurtic)

Normal distribution: Kurtosis of 3 (mesokurtic) --> pandas already calculates kurtosis -3 --> for pandas neutral would be a value of 0

Lower kurtosis: Light tails, fewer outliers (platykurtic)

# Suche nach Mustern in den Features

```python
data_copy = df_clean.copy()

def test_feature_combination(df, feature1, feature2, target='mathe_note'):
    """Testet, ob Merkmalskombinationen stärker mit der Mathenote korrelieren."""
    def get_condition_and_name(f):
        if isinstance(f, tuple):
            return df.eval(f[0]), f[1]
        if df[f].dtype == 'object':
            top_cat = df[f].value_counts().index[0]
            return df[f] == top_cat, f"{f}_{top_cat}"
        else:
            thresh = df[f].median()
            return df[f] > thresh, f"{f}>{thresh}"

    cond1, name1 = get_condition_and_name(feature1)
    cond2, name2 = get_condition_and_name(feature2)
    
    combined_name = f"{name1}_AND_{name2}"
    df[combined_name] = cond1 & cond2
    
    # Korrelationen (Achtung: negative Korrelation bedeutet hier meist 'bessere' Note)
    corr1 = df[target].corr(cond1.astype(int))
    corr2 = df[target].corr(cond2.astype(int))
    corr_combined = df[target].corr(df[combined_name].astype(int))
    
    is_stronger = abs(corr_combined) > max(abs(corr1), abs(corr2))
    df.drop(combined_name, axis=1, inplace=True)
    
    return {'feature1': name1, 'feature2': name2, 'corr1': corr1, 
            'corr2': corr2, 'corr_combined': corr_combined, 'is_stronger': is_stronger}

feature_groups = {
    'Education': ['skill_programmieren', 'skill_englisch', ('mathe_lk == "Ja"', 'mathe_lk_Ja')],
    'Demographics': ['alter_aktuell', ('herkunft_typ == "Stadt"', 'is_stadt'), ('geschlecht == "Weiblich"', 'is_w')],
    'Lifestyle': [('nebenjob == "Ja"', 'hat_job'), 'skill_programmieren']
}

strong_combinations = []
for group_name, features in feature_groups.items():
    print(f"\n=== Testing {group_name} Combinations ===")
    for i in range(len(features)):
        for j in range(i+1, len(features)):
            result = test_feature_combination(data_copy, features[i], features[j])
            if result['is_stronger']:
                strong_combinations.append(result)
                print(f"Treffer: {result['feature1']} & {result['feature2']} (Corr: {result['corr_combined']:.3f})")

top_threshold = df_clean['mathe_note'].quantile(0.15)
bottom_threshold = df_clean['mathe_note'].quantile(0.85)

def identify_extreme_groups(df, feature, target='mathe_note'):
    print(f"\nAnalyse für: {feature}")
    if df[feature].dtype == 'object':
        for cat in df[feature].unique():
            subset = df[df[feature] == cat]
            if len(subset) == 0: continue
            top = (subset[target] <= top_threshold).mean() * 100
            bottom = (subset[target] >= bottom_threshold).mean() * 100
            if top > 25 or bottom > 25:
                print(f"  Gruppe {cat}: Top {top:.1f}%, Bottom {bottom:.1f}% (n={len(subset)})")
    else:
        # Numerisch: Median-Split
        low_half = df[df[feature] <= df[feature].median()]
        top = (low_half[target] <= top_threshold).mean() * 100
        print(f"  Niedriger Wert {feature}: Top-Performer Anteil {top:.1f}%")

# Wichtige Spalten prüfen
for f in ['studiengang', 'mathe_lk', 'skill_programmieren', 'skill_englisch']:
    identify_extreme_groups(df_clean, f)
```
